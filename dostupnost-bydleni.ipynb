```python
import pandas as pd
import matplotlib.pyplot as pl
```

```python
data = pd.DataFrame.from_csv('dostupnost_bydleni.csv', sep=',', index_col=None)
kraje = {1001: 'Praha', 1003: 'Jihočeský', 1004: 'Pardubický', 1005: 'Královéhradecký', 1006: 'Vysočina', 1007: 'Karlovarský', 1008: 'Liberecký', 1010: 'Plzeňský', 1011: 'Středočeský', 1013: 'Ústecký', 2002: 'Jihomoravský', 2009: 'Olomoucký', 2012: 'Moravskoslezský', 2014: 'Zlínský', 0: 'Celá ČR'}
data['kraj'] = data['id_kraj'].map(kraje)
typy = {0: 'všechny byty', 113: '1+1', 123: '2+1', 126: '3+1', 133: '4+1'}
data['byt'] = data['druh'].map(typy)

prijmy = pd.DataFrame.from_csv('prijmy_domacn_2011.csv', sep=',')

najmy = pd.DataFrame.from_csv('ceny_najemneho.csv', sep=',')
```

```python
data.byt.unique()
```

```python
najmy.groupby(['kraj']).agg({'najemne_min (kc/m2/mesic)':{'Min nájemné': 'mean'}}).sort(ascending=True)
```

```python
for najem in najmy['velikostni kategorie'].unique():
    print najem
```

```python
def f(x):
    alfa = x.split("-")
    return alfa[2] + "-" + alfa[3] + "-1"

datum = pd.to_datetime(data['kraj-druh-R-M'].map(f))
data['datum'] = datum
```

```python
data[(data.id_kraj == 1001) & (data.byt == '1+1')].mesic.sort()
```

```python
stamps = data[(data.byt == '1+1') & (data.id_kraj == 1001)]['datum']
cena = data[(data.byt == '1+1') & (data.id_kraj == 1001)]['cena']

cena2 = data[(data.byt == '2+1') & (data.id_kraj == 1001)]['cena']
cena3 = data[(data.byt == '3+1') & (data.id_kraj == 1001)]['cena']
cena4 = data[(data.byt == '4+1') & (data.id_kraj == 1001)]['cena']



pl.plot(stamps, cena, c="red")
pl.plot(stamps, cena2, c="blue")
pl.plot(stamps, cena3, c="green")
pl.plot(stamps, cena4, c="yellow")


pl.show()

#cena.resample('A-jul').plot(kind='line')
```
